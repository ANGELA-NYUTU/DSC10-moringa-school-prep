## 1.4 Cleaning our Dataset

Let us perform data cleaning procedures below providing a documentation for our actions and reasons. We will perform as many data cleaning procedures as we think suitable for the various dimensions of data.

### 1.41) Validity: <font color="green">Challenges</font>

```
# Procedure 1: Irrelevant Data Observation
# Data Cleaning Action: Dropping Location_1 attribute
# Explanation: We won't need it during Analysis. No question to be answered requires that column.
#
df.drop('Location_1', axis = 1, inplace = True)
df.head(10)
```

```
# Procedure 2: Irrelevant Data Observation
# Data Cleaning Action: Drop OBJECTID attribute
# Explanation: Its irrelevant to our table i.e. attribute starts from 0 and ends of 48 i.e. No meaning at all.
# 
df.drop('OBJECTID', axis= 1, inplace= True)
df.head(10)
```

```
# Procedure 3:
# Data Cleaning Action: checking for syntax errors and correcting them for county column
# Explanation:
#
df['Missing_Hand'] = list(map(lambda x: x[:-1], df['Missing_Hand'].values))
df['Missing_Foot'] = list(map(lambda x: x[:-1], df['Missing_Foot'].values))
df['Lame'] = list(map(lambda x: x[:-1], df['Lame'].values))
df['Blind'] = list(map(lambda x: x[:-1], df['Blind'].values))
df['Deaf'] = list(map(lambda x: x[:-1], df['Deaf'].values))
df['Dumb'] = list(map(lambda x: x[:-1], df['Mental'].values))
df['Mental'] = list(map(lambda x: x[:-1], df['Mental'].values))
df['Paralyzed'] = list(map(lambda x: x[:-1], df['Paralyzed'].values))
df['Other'] = list(map(lambda x: x[:-1], df['Other'].values))
```

```
# Procedure 4:
# Data Cleaning Action:
# Explanation: the 'kenya average' row do not represent any county data but overal kenyan data that is no t required in this analysis

df.drop("Kenya average", inplace = True)
df
```

## 1.5 Answering Questions

Let's answer the following questions with our dataset using python.

```
# Challenge 1
# Which county had the highest no. of registered deaf persons?

url1 = '/content/Population_Distribution_of_PLWDkenya.csv'
df2 = pd.read_csv(url1)
df2.head(5)
df3 = df2['Deaf'].max()
df4 = df2[df2['Deaf'] == df3]
df4.groupby(["Deaf"])['County'].sum()


```

```
# Challenge 2
# Which county had the highest no. of registered persons with a missing hand?
df5 = df2['Missing_Hand'].max()
df6 = df2[df2['Missing_Hand'] == df5]
df6.groupby(["Missing_Hand"])['County'].sum()
```

```
# Challenge 3
# Which county had the highest no. of registered persons with a missing foot?
# 
dfmf = df2['Missing_Foot'].max()
dfmf2 = df2[df2['Missing_Foot'] == dfmf]
dfmf2.groupby(["Missing_Foot"])['County'].sum()
```

```
# Challenge 4
# Which county had the highest no. of registered lame persons?
# 
dfl = df2['Lame'].max()
dfl2 = df2[df2['Lame'] == dfl]
dfl2.groupby(["Lame"])['County'].sum()
```

```
# Challenge 5
# Which county had the lowest no. of registered blind persons?
# 
dfb = df2['Blind'].min()
dfb1 = df2[df2['Blind'] == dfb]
dfb1.groupby(["County"])
dfb1[['County', 'Blind']]
```

```
# Challenge 6
# Which county had the highest third no. of registered deaf persons?
dfd3 = df2.groupby(['County']).sum().sort_values('Deaf', ascending = False)
dfd3.head(3)
```

```
# Challenge 7
# In descending order, which top 5 counties had the highest no. of registered dumb persons?
# 
dfd3 = df2.groupby (['County']).sum().sort_values('Dumb', ascending= False)
dfd3.head(5)
```

```
# Challenge 8
# In ascending order, which top 5 counties had the highest no. of registered persons with a mental disability? 
# 
dfme = df2.groupby (['County']).sum().sort_values('Mental', ascending= False)
dfme.head(5)
```

```
# Challenge 9
# Which counties had no registerd blind persons nor deaf persons?
# 
dfbd = df2[(df2['Blind']) > 0 & (df2['Deaf']> 0)]
dfbd['County']
```

```
# Challenge 10
# Which disability was the most registered across all the counties?
disability = ['Missing_Hand',	'Missing_Foot',	'Lame',	'Blind',	'Deaf',	'Dumb',	'Mental',	'Paralyzed',	'Other']
groupy_disability = df2['disability'].groupby('County') 
groupy_disability.sum()
```

```
# Challenge 12
# What was the average no. of registered persons with a disability?
#
df2['Total_Count'].mean()
```

```
# Challenge 13
# Which three counties had least registered persons with disabilities?
#
dfreg = df2.groupby (['County']).sum().sort_values('Total_Count', ascending= True)
dfreg.head(3)
```

```
# Challenge 14
# What was the total no of registered persons with a disability across all counties?
#
df2['Total_Count'].sum()
```

```
# Challenge 15 
# Which top 3 counties has the highest no. of registered persons with a disability?
# 
dfr = df2.groupby (['County']).sum().sort_values('Total_Count', ascending= False)
dfr.head(3)
```
